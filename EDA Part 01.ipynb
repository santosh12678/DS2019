```python
df_after_remove =df.drop([5,10], axis=0)
```

```python
tranposed_df = df_after_remove.transpose()
tranposed_df.head()
```

```python
tranposed_df.drop(["GSDP - CURRENT PRICES (` in Crore)"],axis =1, inplace =True)
```

```python
tranposed_df.head(5)
```

```python
tranposed_df.columns = tranposed_df.columns.map(lambda x,y:str(x)+ '_btc')
```

```python
tranposed_df.drop(['Duration','West Bengal1','All_India GDP'],axis=0, inplace =True)
```

```python
tranposed_df
```

```python
tranposed_df['Average']= tranposed_df.mean(axis=1)
```

```python
tranposed_df.sort_values(by =['Average'],ascending = False,inplace=True)
```

```python
plt.figure(figsize=(10,8), dpi =80, facecolor ='w', edgecolor='k')

barplot = sns.barplot(x=tranposed_df.index, y = tranposed_df['Average'])

barplot.set_xticklabels(tranposed_df.index,rotation=90)
plt.ylabel("Average Growth")
plt.xlabel("States and UT's")
plt.title("Average Growth of States over the Duration 2013-14, 2014-15 and 2015-16")
#plt.yscale('log')
plt.show()
```
