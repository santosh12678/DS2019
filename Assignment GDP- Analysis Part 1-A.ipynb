```python
#Import Packages 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("white")
```

```python
#Read Data from the CSV 
#NOTE:- File should be present at the code location
raw_df = pd.read_csv("ab40c054-5031-4376-b52e-9813e776f65e.csv")
```

```python
#Checking all the data for formatting like all the rows and columns
    
raw_df.head(20)
```

# <b><i>Step 01</i></b><n></n>


<b> Remove the rows: '(% Growth over the previous year)' and 'GSDP - CURRENT PRICES (` in Crore)' for the year 2016-17.
which has index 5 and 10 </b>

```python
#Remove index 5 and 10 row number which has the rows: '(% Growth over the previous year)' and 'GSDP - CURRENT PRICES (` in Crore)' for the year 2016-17.
df_after_remove =raw_df.drop([5,10], axis=0)
```

```python
#Set the Items Description as index so that it will help us while transpose
df_after_remove.set_index('Items  Description', inplace=True)
df_after_remove.head(5)
```

```python
df_after_remove.columns
```

```python
#Transposed the removed df to plot the correct graph
tranposed_df = df_after_remove.transpose()
tranposed_df.head()
```

```python
#Renaming the Header column with Items Description_Duration

transposed_columns =list(tranposed_df.columns) #All the columns list
years = tranposed_df.loc[['Duration']].values  #All the values in duration, return an np array
year_list = years[0].tolist()  # Convert an np array into list

new_columns =[]
for col, year in zip(transposed_columns, year_list):
    
    new_columns.append(col+"_"+year)

#Created new column list with new header (Items Description_Duration)
```

```python
#Apply the new Column to existing one
tranposed_df.columns = new_columns
tranposed_df
```

```python
tranposed_df.drop(['Duration'],axis=0, inplace =True)
```

```python
#Dropped all the 
dropped_GSDP_df = tranposed_df.drop(["GSDP - CURRENT PRICES (` in Crore)_2011-12","GSDP - CURRENT PRICES (` in Crore)_2012-13","GSDP - CURRENT PRICES (` in Crore)_2013-14","GSDP - CURRENT PRICES (` in Crore)_2014-15","GSDP - CURRENT PRICES (` in Crore)_2015-16","(% Growth over previous year)_2012-13"],axis =1)
```

```python
dropped_GSDP_df.head(5)
```

```python
dropped_GSDP_df['Average']= dropped_GSDP_df.mean(axis=1)
```

```python
dropped_GSDP_df
```

```python
dropped_GSDP_df.sort_values(by =['Average'],ascending = False,inplace=True)
```

```python
dropped_GSDP_df.drop(['All_India GDP','West Bengal1'],axis=0, inplace =True)
```

# <b><i>Step 02</i></b>

<b>Calculate the average growth of states over the duration 2013-14, 2014-15 and 2015-16 by taking the mean of the row '(% Growth over previous year)'. Compare the calculated value and plot it for the states. </b>

```python
plt.figure(figsize=(10,8), dpi =80, facecolor ='w', edgecolor='k')

barplot = sns.barplot(x=dropped_GSDP_df.index, y = dropped_GSDP_df['Average'])

barplot.set_xticklabels(dropped_GSDP_df.index,rotation=90)
plt.ylabel("Average Growth")
plt.xlabel("States and UT's")
plt.title("Average Growth of States over the Duration 2013-14, 2014-15 and 2015-16")
#plt.yscale('log')
plt.show()
```
